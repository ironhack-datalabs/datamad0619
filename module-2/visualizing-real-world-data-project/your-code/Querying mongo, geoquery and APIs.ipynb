```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client.companies
```

```python
offices_list = db.first_office.find({"$and": [{"founded_year": {"$gt": 2005}}, {"number_of_employees": {"$gt": 50}}, {"currencies_ok": {"$gt":1000000}}, {"$or":[{"category_code": {"$eq": "web"}}, {"category_code": {"$eq": "mobile"}}, {"category_code": {"$eq": "news"}}, {"category_code": {"$eq": "social"}}, {"category_code": {"$eq": "games_videos"}}, {"category_code": {"$eq": "network_hosting"}}, {"category_code": {"$eq": "ecommerce"}}, {"category_code": {"$eq": "advertising"}}]}]}, {"name":1, "geo_point": 1, "latitude":1, "longitude":1, "currencies_ok":1})
```

```python
offices_list2 = pd.DataFrame(offices_list)
offices_list2
```

```python
off2= offices_list2.copy()
```

```python
off2 = off2[["name", "currencies_ok", "latitude", "longitude", "geo_point"]]
off2.to_json("juans.json", orient="records")
```

```python
df = pd.DataFrame(offices_list2, columns=["latitude", "longitude"])
geo_list = [df.columns.values.tolist()] + df.values.tolist()
geo_list.pop(0)
print(geo_list)
```

```python
#https://docs.mongodb.com/manual/reference/operator/query/near/index.html

def findNear(geopoint, radio_max_meters):
    return db.final_offices.find({
        "geo_point": {
         "$near": {
           "$geometry": geopoint,
           "$maxDistance": radio_max_meters,
         }
       }
    })

#https://developers.google.com/maps/documentation/geocoding/intro
twitter = {"type":"Point","coordinates":[-122.416924,37.776805]}
radio_max_meters = 2000
num_offices = findNear(twitter, radio_max_meters).count()

print(f"Hay {num_offices} oficinas cerca a {radio_max_meters} metros")
print("La mas cercana es")
print(list(findNear(twitter, radio_max_meters)))
```
