```python
# https://api.mongodb.com/python/current/tutorial.html

from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client.companies
```

```python
import pandas as pd
offices = db.companies.find()
```

```python
offices_data = pd.DataFrame(offices)
offices_data.head()
```

```python
at_least_1_office = db.companies.find({
    "offices":{
        "$not":{
            "$size":0
        }
    }
},{"name":1, "offices":1})
```

```python
df = pd.DataFrame(at_least_1_office)
df.head()
```

```python
def getOffices(data):
    data = data['offices']
    # return (len(data),data[0]['latitude'],data[0]['longitude'])

    # Only create the geoJSON object if all geodata is available
    principal = None
    if data[0]['latitude'] and data[0]['longitude']:
        principal = {
            "type":"Point",
            "coordinates":[data[0]['longitude'], data[0]['latitude']]
        }

    return {
        "latitude": data[0]['latitude'],
        "longitude": data[0]['longitude'],
        "geo_point": principal
    }


all_offices = df[["offices"]].apply(getOffices, result_type="expand", axis=1)
```

```python
all_offices.head()
```

```python
df = pd.concat([offices_data,all_offices], axis=1)
df.head()
```

```python
offices_cols = df.drop(["_id", "acquisition", "acquisitions", "alias_list", "blog_feed_url", "blog_url", "competitions", "created_at", "crunchbase_url", "phone_number", "products", "providerships", "relationships", "screenshots", "tag_list", "twitter_username", "updated_at", "video_embeds", "overview", "permalink", "milestones", "ipo", "investments", "founded_month", "founded_day", "external_links", "email_address", "description", "deadpooled_url", "deadpooled_month", "deadpooled_day", "funding_rounds", "image", "homepage_url", "partners"], axis=1)
```

```python
offices_cols.head()
```

```python
offices_no_nulls = offices_cols[offices_cols["deadpooled_year"].isna()==False].index
offices_clean = offices_cols.drop(offices_no_nulls, axis=0)
offices_clean[offices_clean["deadpooled_year"]!="NaN"].index

offices_cleann = offices_clean
```

```python
url="https://api.exchangeratesapi.io/latest"
query_params={"base":"USD","symbols":["CAD","EUR","GBP","SEK"]}

res = req.get(url,params=query_params)
content=res.json()
rates=content["rates"]
```

```python
import re
```

```python
def moneyDollars(s):
    mfloat=float(re.search("(\d*\.\d*|\d+)",s).group(1))
    if s.endswith("M"): mfloat*=1000000
    if s.endswith("B"): mfloat*=1000000000
    if s.endswith("k"): mfloat*=1000
    if s.startswith("C$"): mfloat*=1/rates["CAD"]
    if s.startswith("€"): mfloat*=1/rates["EUR"]
    if s.startswith("£"): mfloat*=1/rates["GBP"]
    if s.startswith("kr"): mfloat*=1/rates["SEK"]
    return round(mfloat)

data_last["currencies_ok"] = data_last.copy()["total_money_raised"].apply(moneyDollars)
data_last.head()
```

```python
data_last["founded_year"] = data_last["founded_year"].astype(int)
```
