```python
# https://api.mongodb.com/python/current/tutorial.html

from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client.companies
```

```python
import pandas as pd
offices = db.companies.find()
```

```python
offices_data = pd.DataFrame(offices)
offices_data.head()
```

```python
offices_cols = offices_data.drop(["_id", "acquisition", "acquisitions", "alias_list", "blog_feed_url", "blog_url", "competitions", "created_at", "crunchbase_url", "phone_number", "products", "providerships", "relationships", "screenshots", "tag_list", "twitter_username", "updated_at", "video_embeds", "overview", "permalink", "milestones", "ipo", "investments", "founded_month", "founded_day", "external_links", "email_address", "description", "deadpooled_url", "deadpooled_month", "deadpooled_day", "funding_rounds", "image", "homepage_url", "partners"], axis=1)
```

```python
offices_no_nulls = offices_cols[offices_cols["deadpooled_year"].isna()==False].index
offices_clean = offices_cols.drop(offices_no_nulls, axis=0)
offices_clean[offices_clean["deadpooled_year"]!="NaN"].index

offices_cleann = offices_clean
```

```python
offices_no_years = offices_cleann[offices_cleann["founded_year"].isna()==True].index
offices_reclean = offices_cleann.drop(offices_no_years, axis=0)
offices_reclean[offices_reclean["founded_year"]!="NaN"].index

offices_extraclean = offices_reclean
```
