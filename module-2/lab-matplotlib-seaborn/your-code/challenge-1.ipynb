#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
x = np.arange(0,100)
y=x**2
z=np.exp(x)
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(x,y,'b-',linewidth=3)
plt.plot(x,z,'r-',markersize=3)
plt.yscale("linear")
plt.subplot(1,2,2)
plt.plot(x,y,'b-',linewidth=3)
plt.plot(x,z,'r-',markersize=3)
plt.yscale("log")
```
