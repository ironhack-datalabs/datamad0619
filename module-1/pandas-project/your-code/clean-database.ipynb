```python

def cleanactivity(act):
    if "Surf" in act or "Body" in act or "Boogie" in act or "surf" in act or "Paddleskiing" in act\
    or "Riding" in act or "Paddle" in act:
        return "Surfing"
    elif "Swimming" in act or "swimming" in act or "Floating" in act or "Tread" in act or "Bath" in act\
    or "bath" in act or "Splash" in act or "Jump" in act or "Playing" in act:
        return "Swimming"
    elif (("Fishing" in act or "fishing" in act) and ("shark" in act or "Shark" in act)) or ("shark" in act and "net" in act):
        return "Fishing sharks"
    elif "Fishing" in act or "fishing" in act or "Shrimping" in act:
        return "Fishing"
    elif "Walk" in act or "walk" in act or "Stand" in act or "Wading" in act:
        return "Walking"
    elif "Diving" in act or "Dive" in act or "diving" in act or "dive" in act:
        return "Diving"
    elif "Canoeing" in act or "Kayak" in act or "Boating" in act or "boat" in act or "Rowing" in act \
    or "rowing" in act or "overboard" in act or "Sea Disaster" in act or "Shipwreck" in act or "ship" in act:
        return "Boat"
    elif "Spearfishing" in act or "Spearing" in act:
        return "Spearfishing"
    else:
        return act
    
wdf = df_cleaning.copy()
wdf['Activity'].fillna("Unknown",inplace=True)
wdf["Activity"].apply(str)
wdf["Activity"].apply(cleanactivity).value_counts()
```
