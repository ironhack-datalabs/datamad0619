```python
#LIMPIEZA DE LA VARIABLE DATE
```

```python
def findmes (m):
    m = str(m)
    if len(re.findall('\-\w+\-',m))==0:
        return '-'
    else:
        t= re.findall('\-\w+\-',m)
        return t[0][1:4]
```

```python
SharknoNa['Date']= SharknoNa['Date'].apply(findmes)
```

```python
#LIMPIEZA AGE
```

```python
def findage (t):
    
    t = str(t)
    b = re.findall ('\d+', t)
    
    if len(b) == 1:
        return b[0]
    else:
        return '-'
```

```python
SharknoNa1['Age']= SharknoNa1['Age'].apply(findage)
```

```python
SharknoNa1=SharknoNa1[SharknoNa1['Age']!='-']
```

```python
#Limpieza de FATAL (Y\N)
```

```python
def findFATAL (m):
    
    m = str(m)
    b = re.findall ('N|Y', m)
    
    if len(b) == 1:
        return b[0]
    else:
        return '-'
```

```python
SharknoNa1['Fatal (Y/N)']= SharknoNa1['Fatal (Y/N)'].apply(findFATAL)
```

```python
SharknoNa1=SharknoNa1[SharknoNa1['Fatal (Y/N)']!='-']
display(SharknoNa1)
```
