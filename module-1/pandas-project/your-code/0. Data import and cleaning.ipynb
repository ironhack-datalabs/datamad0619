```python
# Lo primero importar librerias necesarias:
import pandas as pd
import numpy as np
import matplotlib as mp
```

```python
# He descargado el .zip desde Kaggle y lo he descomprimido. Lo guardo en la misma carpeta del notebook.
#Ahora a importar los datos en raw:

sharks=pd.read_csv('./GSAF5.csv',sep=",",engine='python')
print(f"The df has {sharks.shape[0]} rows and {sharks.shape[1]} columns ")
sharks.head()

#ojo, he tenido que especificar engine='python' porque bueno, estamos trabajando con python
```

```python
#miremos cuantos valores nulos tiene cada columna:
null_cols = sharks.isnull().sum()
print(null_cols)
```

```python
'''
Podemos ver que las dos ultimas columnas practicamente no tienen datos, por lo que podemos eliminarlas de nuestro
DF. Para limpiar vamos a crear una copia.
'''

clean_sharks = sharks.drop(["Unnamed: 22","Unnamed: 23"], axis=1)
clean_sharks.head()
```

```python
'''
al revisar el DF vemos que en la mayorÃ­a de los casos las columnas ["Case Number","Case Number.1","Case Number.2"] contienen
los mismo valores. Solamente 13 registros tienen valores diferentes entre ellos
'''
clean_sharks[((clean_sharks["Case Number"] != clean_sharks["Case Number.1"])|(clean_sharks["Case Number"] != clean_sharks["Case Number.2"]))][["Case Number","Case Number.1","Case Number.2"]]
```

```python
'''
por eso me las quito del df
'''
clean_sharks.drop(["Case Number.1","Case Number.2"], axis=1,inplace=True)
clean_sharks.head()
```

```python
'''
Lo mismo ocurre con las columnas href formula y href. Tras una inspeccion ambas columnas parecene contener
las mismas urls en cada registro. Solo 54 registros tienen diferencias entre ellos y, en la mayorÃ­a de los casos
la url presente en la columna href es correcta mientras que la otra no funciona. Por este motivo eliminamos la 
columna href
'''
print(clean_sharks[(clean_sharks["href formula"] != clean_sharks["href"])][["href formula","href"]].head())
clean_sharks.drop(["href"], axis=1,inplace=True)
clean_sharks.head()
```
