# 1. Importo base de datos limpia y la biblioteca pandas:

```python
import pandas as pd
shark_attack = pd.read_csv("clean_sharks_attack.csv")
display(shark_attack)
```

   # 2. Extraigo las columnas que voy a necesitar para el anÃ¡lisis:

```python
sharks_columns_1= shark_attack[['Species','Fatal (Y/N)','Activity']]
```

```python
display(sharks_columns_1)
```

# 3. Me quedo con los datos de gente que ha muerto por ataques de tiburon eliminando los que no han muerto y los Unknown:

```python
#display(sharks_columns_1)
```

```python
sharks_columns_2= sharks_columns_1.drop(sharks_columns_1[sharks_columns_1['Fatal (Y/N)']=='NO'].index)
display(sharks_columns_2)
```

```python
sharks_columns_DEAD= sharks_columns_2.drop(sharks_columns_2[sharks_columns_2['Fatal (Y/N)']=='Unknown'].index)
display(sharks_columns_DEAD)
```

# 4. Me quedo con las columnas de especies y actividades que han causado muertes:

```python
Species_Activity= sharks_columns_DEAD[['Species','Activity']]
display(Species_Activity)
```

# 5. Hago una matriz con las dos columnas para contar la muertes relacionando cada actividad por cada especie de tiburon: 

```python
Species_Activity=pd.crosstab(Species_Activity.Species, Species_Activity.Activity)
```

```python
display(Species_Activity)
```

## 7.1. Saco el porcentaje de actividades que han acabado en muerte por los ataques de tiburones:

```python
Count_activities=pd.DataFrame(sharks_columns_1['Activity'].value_counts())
display(Count_activities)
```

```python
Count_Activity_dead=pd.DataFrame(sharks_columns_DEAD['Activity'].value_counts())
display(Count_Activity_dead)
```

```python
Count_Activity_dead=Count_Activity_dead.rename(columns={'Activity':'Killer_activities'})
display(Count_Activity_dead)
```

```python
concatenate_activities=pd.concat([Count_activities,Count_Activity_dead], axis=1, sort=False)
display(concatenate_activities)
#result = pd.concat([df1, df4], axis=1, sort=False)
```

```python
concatenate_activities['Percentage'] = (concatenate_activities['Killer_activities'] / concatenate_activities['Activity']) *100


#concatenate_activities.append(Species_Activity.sum().rename('Percentage')).assign(TOTAL=(lambda x:100 * x / float(x.sum())))


#Species_Activity['gender'].value_counts(normalize=True) * 100
#lambda x:100 * x / float(x.sum()))
concatenate_activities
```

```python

display_percentage= concatenate_activities[['Percentage']]
display_percentage.sort_values('Percentage', inplace=True)
display_percentage.plot(y='Percentage', kind='bar', legend=True, figsize=(15,7))

#display_percentage.plot.bar()
```

## 7.2. Saco el porcentaje de tiburones que han causado mas muertes por sus ataques:

```python
Count_sharks=pd.DataFrame(sharks_columns_1['Species'].value_counts())
display(Count_sharks)
```

```python
Killer_sharks=pd.DataFrame(sharks_columns_DEAD['Species'].value_counts())
Killer_sharks=Killer_sharks.rename(columns={'Species':'Dangerous_species'})
display(Killer_sharks)
```

```python
concatenate_sharks=pd.concat([Count_sharks,Killer_sharks], axis=1, sort=False)
display(concatenate_sharks)
```

```python

concatenate_sharks_clean= concatenate_sharks[['Species', 'Dangerous_species']].fillna(0)
display(concatenate_sharks_clean)
```

```python
concatenate_sharks_clean['Percentage'] = (concatenate_sharks_clean['Dangerous_species'] / concatenate_sharks_clean['Species']) *100
display(concatenate_sharks_clean)
```

```python
display_percentage_sharks= concatenate_sharks_clean[['Percentage']]
display_percentage_sharks.sort_values('Percentage', inplace=True)
display_percentage_sharks.plot(y='Percentage', kind='bar', legend=True, figsize=(15,7))
```
