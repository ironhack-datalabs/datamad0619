#### First was grouping all the fishing activities in a same activity called "Fishng and other types"

```python
def traducefish(m):
    a=str(m).lower()
    for word in ["fish", "fishing", "clamming", "lobstering", "crabbing", "hunting lobsters"]:
        if word in a:
            return "fishing and other types"
    return a
```

```python
data_clean['Activity'] = data_clean['Activity'].apply(traducefish)
data_clean["Activity"].value_counts()
```

#### Then, all types of surf description in the surfing activity

```python
def traducesurf(m, lexema="surf", resultado="surfing"):
    if lexema in str(m).lower():
        return resultado
    else:
        return m
```

```python
data_clean['Activity'] = data_clean['Activity'].apply(traducesurf)
data_clean["Activity"].value_counts()
```

```python
data_clean['Activity'] = data_clean['Activity'].apply(traduceswim)
data_clean["Activity"].value_counts()
```

#### As we become more species/animal conscious, we try to help injured or in danger animals and that activity results with a big number of people injured trying to help sharks. Even if the shark is thankfull for the need, any movement can result in an injury. 

```python
def traducehelpshark(m):
    a=str(m).lower()
    for word in ["removing shark from a net", "removing shark from a trap", "caught in net", "injured shark", "stranded", "grounded", "beached", "net", "land a shark"]:
        if word in a:
            return "Helping sharks"
    return a
```

#### The documentation said that sharks are not naturally agressive, the bite when they are curious with an object (can be a human or other) and when they feel threatened. So all attack to a shark will result in a human injury or with an attack in return.

```python
def traduceattack(m):
    a=str(m).lower()
    for word in ["spearing", "pulling shark from the water", "wrangling a shark", "to ride a shark", "harassing", "killing sharks", "killing sharks", "kill shark", "kill sharks", "attack a shark", "attacking a shark", "killing a shark", "chase shark"]:
        if word in a:
            return "Attacking sharks"
    return a
```

```python
data_clean['Activity'] = data_clean['Activity'].apply(traduceattack)
data_clean["Activity"].value_counts()
```

## Time to analyze ð¤

```python
new_df = pd.DataFrame()
new_df['TOTAL'] = data_last[data_last['SEX']!=None].groupby(['COUNTRY', 'ACTIVITY']).count()['SEX']
new_df['MALE %'] = (data_last[data_last['SEX']=="M"].groupby(['COUNTRY', 'ACTIVITY']).count()['SEX'])*100/new_df['TOTAL']
new_df['FEMALE %'] = (data_last[data_last['SEX']=="F"].groupby(['COUNTRY', 'ACTIVITY']).count()['SEX'])*100/new_df['TOTAL']
new_df['AVG AGE MALE'] = (data_last[(data_last['AGE']!=None)&(data_last['SEX']=="M")].groupby(['COUNTRY', 'ACTIVITY']).mean()['AGE'])
new_df['AVG AGE FEMALE'] = (data_last[(data_last['AGE']!=None)&(data_last['SEX']=="F")].groupby(['COUNTRY', 'ACTIVITY']).mean()['AGE'])
new_df['TOTAL FATALS'] = data_last[(data_last['FATAL']=="Y")].groupby(['COUNTRY', 'ACTIVITY']).count()['SEX']
new_df['% FATALS MALE'] = data_last[(data_last['FATAL']=="Y")&(data_last['SEX']=="M")].groupby(['COUNTRY', 'ACTIVITY']).count()['SEX']*100/new_df['TOTAL FATALS']
new_df['% FATALS FEMALE'] = data_last[(data_last['FATAL']=="Y")&(data_last['SEX']=="F")].groupby(['COUNTRY', 'ACTIVITY']).count()['SEX']*100/new_df['TOTAL FATALS']
new_df['AVG AGE MALE FATALS'] = (data_last[(data_last['AGE']!=None)&(data_last['SEX']=="M")&(data_last['FATAL']=="Y")].groupby(['COUNTRY', 'ACTIVITY']).mean()['AGE'])
new_df['AVG AGE FEMALE FATALS'] = (data_last[(data_last['AGE']!=None)&(data_last['SEX']=="F")&(data_last['FATAL']=="Y")].groupby(['COUNTRY', 'ACTIVITY']).mean()['AGE'])
new_df.round(decimals=2).fillna(0)
```
