## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
weather = pd.read_csv('../weather-raw.csv')
display(weather)
```

```python
display(weather.isnull().sum()==22)
```

```python
col_days = list(weather.columns[4:])
print(col_days)
#table = pd.pivot_table(weather, values= ,index=col_days, columns=['element'])
#print(table)

weather_melt=weather.melt(id_vars=weather.columns[:4],value_vars=weather.columns[4:])
display(weather_melt.head())
weather_melt.shape
```

```python
weather_pivot = pd.pivot_table(weather_melt, values= 'value',index=['id', 'year', 'month','variable'], columns=['element'])
weather_pivot = weather_pivot.reset_index().rename_axis(None).rename_axis(None, axis=1)
display(weather_pivot.head())
```

```python
weather_pivot.to_csv('weather_pivot.csv', sep=',', index=False)
```
