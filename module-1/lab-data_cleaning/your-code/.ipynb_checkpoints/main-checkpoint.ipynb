#### 4. Import the users table 

```python
sqlquery = """
SELECT * FROM users
"""

users = pd.read_sql_query(sqlquery, engine)
print(users.shape)
users.head()
```

#### 5. Rename Id column to userId

```python
users.rename(columns={"Id":"userId"},inplace=True)
users.head()
```

#### 6. Import the posts table. 

```python
sqlquery1="""SELECT * FROM posts"""

posts = pd.read_sql_query(sqlquery1, engine)
print(posts.shape)
posts.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
posts.rename(columns={"Id":"postId","OwnerUserId":"userId"},inplace=True)
posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
users_1=users[["userId","Reputation","Views","UpVotes","DownVotes"]]
posts_1=posts[["postId","Score","userId","ViewCount","CommentCount"]]
users_1.head()
```

```python
posts_1.head()
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
up_merge=users_1.merge(posts_1,how="inner",on="userId")
up_merge.head()
#I've used an inner join so that the merged table only takes into account posts written by existing users 
#and users 
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
null_cols = up_merge.isnull().sum()
print(null_cols[null_cols > 0])
print("\nshape: ",up_merge.shape)
```

```python
"""
The ViewCount is null in more than half of our rows, which would suggest that the ViewCount system isn't working 
properly. It will be necessary to have a look at it. Until then, we will replace null values for zeroes to make sure
we can spot these later (the minimum ViewCount is 1.0, so it won't be difficult)
"""
up_merge["ViewCount"].fillna(0,inplace=True)
up_merge.head()
```

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
stats = up_merge.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
outliers = pd.DataFrame(columns=up_merge.columns)

for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = up_merge[(up_merge[col] < lower) | 
                   (up_merge[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
```

```python
outliers.head()
```

```python
#to csv:
outliers.to_csv("./outliers.csv", sep=",",index=False)
```
