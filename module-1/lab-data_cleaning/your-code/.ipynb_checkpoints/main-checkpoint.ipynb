#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data 


```python
import getpass
import pymysql
from sqlalchemy import create_engine
```

#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/search?tableCount%5B%5D=0-10&tableCount%5B%5D=10-30&dataType%5B%5D=Numeric&databaseSize%5B%5D=KB&databaseSize%5B%5D=MB)

```python
mysql_domain = "relational.fit.cvut.cz"
mysql_user = "guest"
mysql_password = getpass.getpass()
mysql_db = "stats"

dbdata = [mysql_user, mysql_password, mysql_domain, mysql_db]
connection_url = 'mysql+pymysql://{}:{}@{}/{}'.format(*dbdata)
engine = create_engine(connection_url)
print("Connected to DB!")
```

#### 4. Import the users table 

```python
sqlquery = """
SELECT * FROM users
"""

data = pd.read_sql_query(sqlquery, engine)
print(data.shape)
data.head()
```

#### 5. Rename Id column to userId

```python
d_users = data.rename (columns= {'Id': 'userId'})
d_users.head()
```

#### 6. Import the posts table. 

```python
sqlquery = """
SELECT * FROM posts
"""

dposts = pd.read_sql_query(sqlquery, engine)
dposts.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
d_posts = dposts.rename (columns= {'Id':'postId','OwnerUserId': 'userId'})
d_posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
new_users = d_users[['userId','Reputation','Views','UpVotes','DownVotes']]
new_users.head()
```

```python
new_posts= d_posts[['postId','Score','userId','ViewCount','CommentCount']]
new_posts.head()



```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
users_posts = new_users.merge(new_posts, left_on= 'userId', right_on= 'userId')
users_posts.head()
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
users_posts.isnull().sum()
```
